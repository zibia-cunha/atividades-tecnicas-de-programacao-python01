    Crie um filtro para Kevin e Laura.
    Crie um filtro para todos que tiveram um numero de tentativa(attempts) mair que 2.

```python
exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
    }
```

```python
df = pd.DataFrame(exam_data)
df
```

```python
df_reset = df.reset_index()
df_reset
```

```python
filtro = df_reset[df_reset['name'].isin(['Kevin', 'Laura'])]
print(filtro)
```

```python
tentativas_maior_2 = df[df['attempts'] > 2]
tentativas_maior_2
```

```python
df = df.fillna(0)
df
```

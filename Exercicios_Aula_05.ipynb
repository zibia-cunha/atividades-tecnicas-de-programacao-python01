
Bora praticar!
Utilizando o DataFrame importado anteriormente (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)



```python
df_notas = pd.read_csv('alunos3.csv', sep=';')
df_notas
```

```python
colunas = ['Prova_1', 'Prova_2', 'Prova_3']
df_notas[colunas] = df_notas[colunas].apply(lambda x: x.str.replace(',', '.').astype(float))
df_notas
```

```python
df_notas['Prova_4'] = df_notas['Prova_4'].astype(float)
df_notas
```

```python
df_notas['Media_provas'] = df_notas[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']].mean(axis=1)
df_notas
```

Quem foram os alunos que obtiveram a maior e a menor média

```python
maior_media = df_notas['Media_provas'].max()
menor_media = df_notas['Media_provas'].min()
print(f'Maior média foi do aluno José, com: {maior_media} \nMenor média foi do aluno Joao, com: {menor_media}')
```

Agora una este dataframe com o cadastro_alunos.xlsx

```python
df_cadastro = pd.read_excel('cadastro_alunos.xlsx')
df_cadastro
```

```python
df_total = pd.merge(df_notas, df_cadastro, on='RA')
df_total
```

#### Qual a média entre as Media_provas dentro do público feminino? e masculino?






```python
media_por_sexo = df_total.groupby('Sexo')['Media_provas'].mean()
media_por_sexo
```

 Qual a média de idade das pessoas que obtiveram Media_provas maior ou igual a 7?



```python
df_maior_que_7 = df_total[df_total['Media_provas'] >= 7]
df_maior_que_7
```

```python
media_idade = df_maior_que_7['Idade'].mean()
media_idade
```

Qual das cidades possui o maior média de Media_provas? E qual é este valor?



```python
media_por_cidade = df_total.groupby('Cidade')['Media_provas'].mean()
media_por_cidade
```

```python
cidade_maior_media = media_por_cidade.idxmax()
cidade_maior_media
```

```python
valor_maior_media = media_por_cidade.max()
valor_maior_media
```
